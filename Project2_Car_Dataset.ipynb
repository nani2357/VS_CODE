```python
#Missing values: This shows if there are any missing values in the dataset and how many are there in each column.
# We have missing values we need to clean the data
df.isnull().sum()
```

```python
null_rows = df[df.isnull().any(axis=1)]
null_rows
```

```python
# using dropna() also can use df = df[df['Invoice', 'MSRP', 'Make'].notna()]

df.dropna(subset=['Invoice', 'MSRP', 'Make'], inplace = True)
```

```python
df.isnull().sum()
```

```python
# we fill cylinder column with mean, median or mode.
# have check data to which is sutiable
# here cylinder is distant value, we replace more repated value 'mode'

df['Cylinders'].mode()[0]
```

```python
# Fill null values with mode .fillna()
df['Cylinders'].fillna(df['Cylinders'].mode()[0], inplace = True)
```
