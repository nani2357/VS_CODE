# Udemy Dataset

The commands that we used in this project :

* import pandas as pd -- To import Pandas library.
* pd.read_csv - To import the CSV file in Jupyter notebook.
* head() - It shows the first N rows in the data (by default, N=5).
* unique( ) - It shows the all unique values of the column.
* value_counts - In a column, it shows all the unique values with their count. It can be applied to a single column only.
* df[df.Col_1 = = ‘Element1’] - Filtering – We are accessing all records with Element1 only of Col_1.
* df.sort_values(‘Col_name' ,  ascending=False ) - To sort the dataframe wrt any column values in descending order.
* df[ (df.Col1 = = ‘Element1’) & (df.Col2 == ‘Element2’) ] - Multilevel filtering - And Filter – Filtering the data with two & more items.
* str.contains('Value_to_match’) - To find the records that contains a particular string.
* dtypes - To show datatypes of each column.
* pd.to_datetime(df.Date_Time_Col) - To convert the data-type of Date-Time Column into datetime[ns] datatype.
* dt.year - Creating a new column with only year values.
* df.groupby(‘Col_1’)['Col_2'].max() - Using groupby with two different columns.

```python
#Q. 1) What are all different subjects for which Udemy is offering courses ?
# using unique() or value.counts()

df.subject.unique()
```

```python
#  2) Which subject has the maximum number of courses.


courses_offered = df.subject.value_counts()
courses_offered
```

```python
#  3) Show all the courses which are Free of Cost

free_courses = df[df.is_paid == False]
free_courses
```

```python
# 4) Show all the courses which are Paid.

paid_course = df[df.is_paid == True]
paid_course
```

```python
# Q. 5) Which are  Top  10 Selling Courses ?

top_selling = df.sort_values('num_subscribers', ascending=False).head(10)
top_selling
```

```python
#Which are Least Selling Courses ?
least_selling = df.sort_values('num_subscribers', ascending=True).head(10)
least_selling
```

```python
# Q. 7) Show all courses of Graphic Design where the price is below 100 ?

# convert price column to intg

df['price'] = pd.to_numeric(df['price'], errors='coerce') 
# errors='coerce' convert non numerical value to NaN
# drop NaN values or fill with mean value
```

```python
df.isnull().sum()
```

```python
null_rows = df[df.isnull().any(axis=1)]
null_rows
```

```python
null_rows[null_rows['is_paid'] == False]
```

```python
# is paid values for NaN are False, so all course price is 0 and free
# we replace NaN with 0
# using fillna()

df['price'] = df['price'].fillna(0) 

```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['price'] = df['price'].astype(int)
```

```python
# Now dtypes got changed 
#Q. 7) Show all courses of Graphic Design where the price is below 100 ?
df.subject.value_counts()
```

```python
graphic_design = df[df.subject == 'Graphic Design']
graphic_design
```

```python
graphic_design_price_below_100 = graphic_design[graphic_design.price < 100]
graphic_design_price_below_100
```

```python
# Q. 8) List out all the courses that are related to 'Python'.
# using str.contains()fun and case=false and regex=true

courses_python = df[df.course_title.str.contains('python', case=False, regex=True)]
```

```python
courses_python.head(2)
```

```python
#9) What are courses that were published in the year 2015 ?
# check data types
# create new column year
# filter fuction

df.dtypes
```

```python
df['published_timestamp'] = pd.to_datetime(df.published_timestamp)
```

```python
df['year'] = df.published_timestamp.dt.year
```

```python
df.head(2)
```

```python
#9) What are courses that were published in the year 2015 ?

courses_2015 = df[df.year == 2015]
courses_2015.head(2)
```

```python
#10) What is the Max. Number of Subscribers for Each Level of courses ?
max_num_subscibers = df.groupby('subject')['num_subscribers'].sum()
max_num_subscibers
```
